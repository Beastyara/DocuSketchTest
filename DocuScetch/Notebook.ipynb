```python
# Firstly, we downloaded and converted json file to python dictionary
```

```python
import json, urllib
url = "https://ai-process-sandy.s3.eu-west-1.amazonaws.com/purge/deviation.json"
response = urllib.request.urlopen(url)
data = json.loads(response.read())
print('Type of data downloaded from URL:')
print(type(data))
print()
print('DATA overview:')
print(data)
```

```python
#Secondly, we path this dictionary to our function draw_plots
#Our function returns list with paths to all saved plots 
```

```python
import draw_plots as d
plotpaths = d.draw_plots(data)
print('The list of paths to the saved plots:')
print(plotpaths)
```

```python
#Lastly, by using matplotlib and cv2, I simply visualize all saved plots 
```

```python
import matplotlib.pyplot as plt
import cv2 as cv
for plotpath in plotpaths:
    plt.imshow(cv.imread(plotpath))
    plt.show()
```
